```python
from PIL import Image
import base64
import io

archivoEntrada = r"<<imagen>>"
archivoSalida = r"<<imagen_base64>>"

# Cargar imagen y reducir tamaño
img = Image.open(archivoEntrada)
img = img.resize((386, 520))  # Reducir resolución
buffered = io.BytesIO()
img.save(buffered, format="JPEG", quality=24)  # Comprimir imagen

# Convertir a base64
img_base64 = base64.b64encode(buffered.getvalue()).decode()
data_uri = f"data:image/jpeg;base64,{img_base64}"

# Guardar data_uri en un archivo txt
with open(archivoSalida, "w") as f:
    f.write(data_uri)
```
